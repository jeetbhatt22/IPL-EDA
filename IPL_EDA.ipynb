```python
team_score = score.groupby(['match_id', 'inning'])['total_runs'].sum().unstack().reset_index()
team_score.columns = ['match_id', 'Team1_score', 'Team2_score', 'Team1_superover_score', 'Team2_superover_score']
```

```python
matches_agg = pd.merge(match, team_score, left_on = 'id', right_on = 'match_id', how = 'outer')
```

```python
matches_agg.head()
```

### Does wining the toss actully helps team win the game accross seasons?

```python
toss = matches_agg.groupby(['season', 'toss_winner']).winner.value_counts().reset_index(name = 'count')
toss['result'] = np.where(toss.toss_winner == toss.winner, 'won', 'lost')
toss_result = toss.groupby(['season', 'toss_winner','result'])['count'].sum().reset_index()

for x in range(2008, 2017, 1):
    toss_result_x = toss_result[toss_result['season'] == x]
    plot = sns.barplot(x="toss_winner", y="count", hue="result", data=toss_result_x)
    plot.set_title('Matches won/lost by teams winning toss \nSeason ' +str(x))
    plot.set_xticklabels(toss_result_x['toss_winner'],rotation='vertical')
    plt.show()
    x+=1
```

### Is wining toss actually wins the game?

```python
match['tosswin_is_matchwin']='no'
match['tosswin_is_matchwin'].loc[match.toss_winner == match.winner]='yes'

tossmwin = match.tosswin_is_matchwin.value_counts()
labels3 = np.array(tossmwin.index)

size3 = np.array((tossmwin/tossmwin.sum())*100)

plt.figure(figsize=(10,8))
plt.pie(size3,labels=labels3,colors=['#FF5733','#76D7C4'],autopct='%1.1f%%',shadow=True, startangle=90)
plt.title("Wining toss equals match win")
plt.show()
```

Not so it is almost 50-50

### Batsman with most number of runs in IPL

```python
mostruns = score.groupby('batsman')['batsman_runs'].agg(sum).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
mostruns= mostruns.iloc[:10,:]
labels4 = np.array(mostruns['batsman'])
size4 = np.arange(len(labels4))

width= 0.9
fig, ax= plt.subplots(figsize=[10,8])
runs = ax.bar(size4, np.array(mostruns['batsman_runs']), width=width, color='orange')
ax.set_xticks(size4)
ax.set_xticklabels(labels4, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Most Numbers of runs in IPL")
plt.show()
```

Suresh Raina is leading the chart followed by Virat kohli

### Batsman with most number of 4's in their bucket list

```python
boundrys = score.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
boundrys= boundrys.head(10)
label5 = np.array(boundrys['batsman'])
size5 = np.arange(len(label5))

width = 0.9
fig, ax = plt.subplots(figsize=[10,8])
fours =ax.bar(size5, np.array(boundrys['batsman_runs']),width=width, color='blue')
ax.set_xticks(size5)
ax.set_xticklabels(label5,rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most 4's")
plt.show()
```

We can see that Gautam gambhir has highest number of 4's like 480 followed by suresh raina

### Batsman with most number of 6's in their bucket list

```python
sixes = score.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending = False).reset_index(drop=True)
sixes = sixes.head(10)

label6 = np.array(sixes['batsman'])
size6 = np.arange(len(label6))

width = 0.9
fig, ax = plt.subplots(figsize=[10,8])
six =ax.bar(size6, np.array(sixes['batsman_runs']),width=width, color='blue')
ax.set_xticks(size6)
ax.set_xticklabels(label6,rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most 6's")
plt.show()
```

Wooh No Surprise Chris Gayle leads the chart with highest number of sixes followed by suresh raina and rohit sharma

### Top Bowlers over seasons

```python
score["player_dismissed"] = score["player_dismissed"].fillna(0)
#bowler_wickets = score.groupby('bowler')['player_dismissed'].sum()
#bowler_wickets.sort_values(ascending = False, inplace = True)
topbowlers = score.groupby('bowler')['player_dismissed'].agg(lambda x: (x!=0).sum()).reset_index().sort_values(by='player_dismissed', ascending = False).reset_index(drop = True)
topbowlers = topbowlers.head(10)

label8 = np.array(topbowlers['bowler'])
size8 = np.arange(len(label8))
width = 0.9
fig, ax = plt.subplots(figsize=[10,8])
six =ax.bar(size8, np.array(topbowlers['player_dismissed']),width=width, color='Red')
ax.set_xticks(size8)
ax.set_xticklabels(label8,rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers over seasons")
plt.show()
```

We can see from the L. Malinga has taken most no of wickets across all ipl season followed by amit mishra
